```python
'''
Leetcode 1768: Merge Strings Alternatively

You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1. If a string is longer than the other, append the additional letters onto the end of the merged string.

Return the merged string.
'''

def mergeAlternately(word1: str, word2: str) -> str:
    output = ""
    # If we try to access indices higher than min, we'll get errors
    min_len = min(len(word1), len(word2))
    # Add the letter at i = 0, 1, etc. 
    for i in range(min_len):
        output += word1[i]
        output += word2[i]
    # Add the rest of the letters starting from where the iterator stopped (min_len)
    if len(word1) > min_len:
        output += word1[min_len:]
    else:
        output += word2[min_len:]
    return output

word1 = "abcdef"
word2 = "xyz"
expected = "axbyczdef"
print(f"word1 = {word1}\nword2 = {word2}\nexpected = {expected}\n")
print(f"Result of mergeAlternately(): {mergeAlternately(word1, word2)} ✅")
```
